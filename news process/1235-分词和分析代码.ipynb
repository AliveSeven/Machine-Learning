```python
# 这里是逆向最大匹配分词算法
import re

print("--------------------对新闻进行逆向最大匹配分词，分词后的结果是↓--------------------")
ans_reverse=[]  #存放逆向匹配的结果，相当于栈
max_length=45    #分词字典中最大长度字符串的长度，世界上最长的英文单词为45
dic_list=[] #分词字典列表
NiXiang = [] # 存放分词结果，后面用于错误分析

with open('1235字典.txt',encoding='UTF-8') as fp:  # 对字典进行切分，将里面的元素增加到一个列表里面
    for line in fp:
        line = re.sub(r'[\n]','',line) # 将line中的'\n'去掉
        if(line == ''):  # 去掉''
            continue
        dic_list.append(line)  # 加入分词列表


txtFiles = '1235-预处理.txt'     # 待分词的文本（新闻）
with open(txtFiles, encoding='ANSI') as fp:  # 打开新闻，这里是我的新闻是ANSI，所以以ANSI的方式打开
        for line in fp:
            sentence = re.sub(r'[\n]','',line) #待分词的句子，去除\n
            #2、逆向最大匹配
            len_row=len(sentence)          #len_orw为当前为划分句子的长度
            while len_row>0 :              #当前待划分句子长度为0时，结束划分
                divide = sentence[-max_length:]   # 截取长度为max_length的字符串
                while divide not in dic_list:     # #当前截取的字符串不在分词字典中，则进循环
                    if len(divide)==1:            # #当前截取的字符串长度为1时，说明分词字典无匹配项
                        break                     # #直接保留当前的一个字
                    divide=divide[1:]   #注意这里缩短截取字符串是缩短前部分，保留后部分
                ans_reverse.append(divide)        #记录下当前截取的字符串
                sentence = sentence[0:len(sentence)-len(divide)]      #截取未分词的句子
                len_row = len(sentence)              #记录未分词的句子的长度
        for i in ans_reverse:  # 输出分词结果
            if(i == ' '): # 去掉空格词
                continue
            else:
                print(i)
                NiXiang.append(i) # 将分词去掉空格加到列表里面
```

```python
# 这里是HanLP分词
import re
from pyhanlp import * # 导入要分词用的pyhanlp库

HanList = [] # 创建一个列表保存HanLP分词结果
print("--------------------对新闻进行HanLP分词，分词后的结果是↓--------------------")
print('单词↓↓ 词性↓↓')
txtFiles = '1235-预处理.txt'         # 要分词的文本（新闻）
with open(txtFiles, encoding='ANSI') as fp: 
        for line in fp: 
            line = re.sub(r'[\n]','',line) # 将line中的'\n'去掉
            # 进行HanLP中文分词
            for term in HanLP.segment(line):
                if(term.word == ' '): # 去掉空格词
                    continue
                else:
                    # 输出分词的单词，词性
                    print("%s     %s" % (term.word, term.nature))
                    HanList.append(term.word) # 将HanLP分词结果保存在一个列表里面，后面做错误分析有用
```

```python
# 这里是jieba分词
import re
import jieba
import jieba.posseg as pseg

jiebaList = [] # 创建一个列表保存jieba分词结果
print("--------------------对新闻进行jieba分词，分词后的结果是↓--------------------")
print('单词↓↓ 词性↓↓')

txtFiles = '1235-预处理.txt'  # 待处理的文本（新闻）
with open(txtFiles, encoding='ANSI') as fp:
        for line in fp:
            line = re.sub(r'[\n]','',line) # 将line中的'\n'去掉
            # jieba分词处理
            # cut_all为False，为精准模式，为True是准确模式
            new = jieba.cut(line,cut_all=False)  # 精准分词
            all = ''.join(new)   # new 是generator object，将其变成字符串
            words = pseg.cut(all)    # 词性标注
            for word, flag in words:
                if(word == ' '):   # 去掉空格词
                    continue
                # 分词结果，单词 词性
                print('%s     %s' % (word, flag))
                jiebaList.append(word) # 将jieba分词结果保存在一个列表里面，后面做错误分析有用
```
