```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# X中存储交通工具的参数
# 总长度（米）、时速（km/h）、重量（吨）、座位数量
X = [[96, 85, 120, 400],        # 普通火车
      [144, 92, 200, 600],
      [240, 87, 350, 1000],
      [360, 90, 495, 1300],
      [384, 91, 530, 1405],
      [240, 360, 490, 800],      # 高铁
      [360, 380, 750, 1200],
      [290, 380, 480, 960],
      [120, 320, 160, 400],
      [384, 340, 520, 1280],
      [34, 918, 77, 180],      # 飞机
      [35, 1120, 170.5, 185],
      [40, 785, 230, 240],
      [38, 940, 150, 195],
      [45, 920, 275, 275],
      [76, 1050, 575, 490]]
# y中存储类别，0表示普通火车，1表示高铁，2表示飞机
y = [0]*5+[1]*5+[2]*6

# labels中存储对应的交通工具名称
labels = ('普通火车', '高铁', '飞机')

# 创建并训练模型,设置参数为5
knn = KNeighborsClassifier(n_neighbors=5, weights='distance')
knn.fit(X, y)

unknown = []
uny = []
# 以火车、高铁、飞机的顺序生成60组测试数据集
for i in range(20):
      # 以普通火车的合理范围,随机生成长度、速度、重量、座位变量
      long0 = round(np.random.uniform(80,450),1)
      speed0 = np.random.randint(80,110)
      weight0 = round(np.random.uniform(100,550),1)
      seat0 = np.random.randint(300,1500)
      # 增加火车
      unknown.append([long0,speed0,weight0,seat0])
      uny.append(0)

      # 以普通高铁的合理范围,随机生成长度、速度、重量、座位变量
      long1 = round(np.random.uniform(120,500),1)
      speed1 = np.random.randint(300,500)
      weight1 = round(np.random.uniform(100,850),1)
      seat1 = np.random.randint(300,1400)
      # 增加高铁
      unknown.append([long1,speed1,weight1,seat1])
      uny.append(1)

      # 以普通飞机的合理范围,随机生成长度、速度、重量、座位变量
      long2 = round(np.random.uniform(30,100),1)
      speed2 = np.random.randint(700,1300)
      weight2 = round(np.random.uniform(60,650),1)
      seat2 = np.random.randint(100,600)
      # 增加飞机
      unknown.append([long2,speed2,weight2,seat2])
      uny.append(2)

print("测试样本:",unknown)

# 对未知样本进行分类
result = knn.predict(unknown)
print("测试模型得分:",knn.score(unknown,uny))
for para, index in zip(unknown, result):
    print("测试组:",para, labels[index], sep=':')
```
