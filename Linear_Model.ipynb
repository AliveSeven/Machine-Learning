# 线性回归算法验证

```python
# 导入线型模型模块
from sklearn import linear_model
# 快速操作结构数组的工具
import numpy as np
# 可视化绘制
import matplotlib.pyplot as plt
# 导入dataset模型
from sklearn.datasets import make_regression
# 创建线型回归模型
reg1 = linear_model.LinearRegression()
# X,y各取二十个数据集
X, y = make_regression(n_samples=20, n_features=1, noise=2, random_state=1)

# 输出X和y的数据
print("X:",X)
print("y:",y)

# 将x和y进行拟合操作
reg1.fit(X, y)

# 输出斜率，回归系数
print("回归系数:",reg1.coef_)

# 输出截距，以下画线结束
print("线性截距:",reg1.intercept_)

# 使用训练得到的估计器对输入为X的集合进行预测
pre1 = reg1.predict(X)

# 预测结果
print("使用模型预测的结果:",pre1)

# 绘制散点图 参数：x横轴 y纵轴
plt.scatter(X, y)
plt.plot(X, pre1,c='r')
# 绘制x轴和y轴坐标
plt.xlabel("x")
plt.ylabel("y")
# 显示图形
plt.show()
```

# 岭回归算法验证

```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn import linear_model
# 可视化绘制
import matplotlib.pyplot as plt
# 创建岭回归模型
from sklearn.linear_model import Ridge
# X,y各取二十个数据集
X, y = make_regression(n_samples=20, n_features=1, noise=2, random_state=1)
# 输出X和y的数据
print("X:",X)
print("y:",y)

# 假设alpha = 0.1
regs = Ridge(alpha = 0.1)
regs.fit(X,y)
pres = regs.predict(X)
# 使用训练得到的估计器对输入为X的集合进行预测
print("使用模型预测的结果:",pres)

# 设置20次循环以改变alpha的值(约束项系数,其数值越大，特征对结果的影响越小)
for i in range(20):
    # 随机取值
    reg2 = Ridge(alpha = np.random.randint(i,100))
    # 拟合
    reg2.fit(X,y)
    # 使用训练得到的估计器对输入为X的集合进行预测
    pre2 = reg2.predict(X)
    #绘制散点图 参数：x横轴 y纵轴
    plt.scatter(X,y,c=y)
    plt.plot(X, pre2,c='r')
#绘制x轴和y轴坐标
plt.xlabel("x")
plt.ylabel("y")
# 显示图形
plt.show()
```

# Lasso回归算法验证

```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn import linear_model
# 可视化绘制
import matplotlib.pyplot as plt
# 创建Lasso回归模型
from sklearn.linear_model import Lasso
# X,y各取二十个数据集
X, y = make_regression(n_samples=20, n_features=1, noise=2, random_state=1)
# 输出X和y的数据
print("X:",X)
print("y:",y)

# 假设alpha = 0.1
regs = Lasso(alpha = 0.1)
regs.fit(X,y)
pres = regs.predict(X)
# 使用训练得到的估计器对输入为X的集合进行预测
print("使用模型预测的结果:",pres)

# 设置20次循环以改变alpha的值
for i in range(20):
    reg3 = Lasso(alpha = np.random.randint(i,100))
    reg3.fit(X,y)
    # 使用训练得到的估计器对输入为X的集合进行预测
    pre3 = reg3.predict(X)
    #绘制散点图 参数：x横轴 y纵轴
    plt.scatter(X,y,c=y)
    plt.plot(X, pre3,c='r')
#绘制x轴和y轴坐标
plt.xlabel("x")
plt.ylabel("y")
# 显示图形
plt.show()
```
